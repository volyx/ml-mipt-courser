```python
import math
import numpy as np
import scipy.linalg as linalg
import matplotlib.pyplot as plt
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
a = np.array([[1, 1],[1, 15]])
b = np.array([[f(1)], [f(15)]])
w1 = linalg.solve(a,b)

def f1(w, x):
#     print w
#     print w[0][0], w[1][0]
    return w[0][0] + w[1][0] * x

a = np.array([[1, 1, 1],[1, 8, 64], [1, 10, 100]])
b = np.array([[f(1)], [f(8)], [f(10)]])
w2 = linalg.solve(a,b)

def f2(w, x):
    return w[0][0] + w[1][0] * x + x*x*w[2, 0] 

a = np.array([[1, 1, 1 ,1 ],[1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([[f(1)], [f(4)], [f(10)], [f(15)]])
w3 = linalg.solve(a,b)

def f3(w, x):
    return w[0][0] + w[1][0] * x + x*x*w[2, 0] + x*x*x*w[3, 0] 

t = np.arange(1, 15, 0.1)
s1 = list()
s2 = list()
s3 = list()
s4 = list()
for v in t:
    s1.append(f(v))
    s2.append(f1(w1, v))
    s3.append(f2(w2, v))
    s4.append(f3(w3,v))

plt.plot(t, s1, 'b', t, s2, 'r', t, s3, 'g', t, s4, 'y')

plt.grid(True)
plt.savefig("test.png")
plt.show()


print w4
```
